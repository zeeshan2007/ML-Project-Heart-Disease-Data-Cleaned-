```python
df_final = df_cleaned.copy()

# patient_id drop
df_final.drop(columns=['patient_id'], inplace=True, errors='ignore')

# sex / is_male handle
if 'sex' in df_final.columns:
    df_final['sex'] = df_final['sex'].replace({'Male': 1, 'Female': 0}).astype(int)
    df_final.rename(columns={'sex': 'is_male'}, inplace=True)

elif 'is_male' in df_final.columns:
    print("Column 'sex' already renamed to 'is_male'.")

else:
    print("Neither 'sex' nor 'is_male' column found.")

# exercise_induced_angina
if 'exercise_induced_angina' in df_final.columns:
    df_final['exercise_induced_angina'] = df_final['exercise_induced_angina'].astype(int)
    df_final.rename(columns={'exercise_induced_angina': 'has_exercise_induced_angina'}, inplace=True)

# family_history
if 'family_history' in df_final.columns:
    df_final['family_history'] = df_final['family_history'].astype(int)
    df_final.rename(columns={'family_history': 'has_family_history'}, inplace=True)

# wearable_owner
if 'wearable_owner' in df_final.columns:
    df_final['wearable_owner'] = df_final['wearable_owner'].astype(int)
    df_final.rename(columns={'wearable_owner': 'has_wearable'}, inplace=True)

print(df_final.head())
print(df_final.columns.tolist())
```
